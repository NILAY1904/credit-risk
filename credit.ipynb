```python
plt.figure(figsize = (18,18))
sns.heatmap(df.corr(), annot = True, cmap = "RdYlGn")

plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set storing  all the names of correlated columns having corr more than threshold
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: ####corr value can also be negative so taking abs value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
correlation(df,0.8)
```

Above are the columns which seems to be highly corrleated with each other and have pearson correlation value more than 0.8

# Fitting model using different classification algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import KFold
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier

from sklearn.model_selection import train_test_split
```

```python
X1=X.copy()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X1 = pd.DataFrame(sc_X.fit_transform(X1),columns=X.columns)
```

Scaling is required for logistic regression and SVM.

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,train_size=0.75)
```

```python
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,Y,train_size=0.75)
```

```python
lr=LogisticRegression(random_state=0)
lr.fit(X1_train,y1_train)
y_pred2=lr.predict(X1_test)
print(confusion_matrix(y1_test,y_pred2))
print(accuracy_score(y1_test,y_pred2))
print(classification_report(y1_test,y_pred2))
```

```python
clasn=GaussianNB()
clasn.fit(X_train, y_train)
y_predn=clasn.predict(X_test)
print(confusion_matrix(y_test,y_predn))
print(accuracy_score(y_test,y_predn))
print(classification_report(y_test,y_predn))
```

```python
clas=SVC(kernel='rbf',random_state=0)
clas.fit(X1_train, y1_train)
y_preds = clas.predict(X1_test)
print(confusion_matrix(y1_test,y_preds))
print(accuracy_score(y1_test,y_preds))
print(classification_report(y1_test,y_preds))
```

```python
dt= DecisionTreeClassifier(random_state=0)
dt.fit(X_train,y_train)
y_predd=dt.predict(X_test)
print(confusion_matrix(y_test,y_predd))
print(accuracy_score(y_test,y_predd))
print(classification_report(y_test,y_predd))
```

```python
rf=RandomForestClassifier(random_state=0)
rf.fit(X_train,y_train)
y_predr=rf.predict(X_test)
print(confusion_matrix(y_test,y_predr))
print(accuracy_score(y_test,y_predr))
print(classification_report(y_test,y_predr))
```

```python
classifier = XGBClassifier(random_state=0)
classifier.fit(X_train,y_train)
y_predx = classifier.predict(X_test)
print(confusion_matrix(y_test,y_predx))
print(accuracy_score(y_test,y_predx))
print(classification_report(y_test,y_predx))
```

```python
lr=LogisticRegression()
grid={'C':10.0 **np.arange(-2,3),'penalty':['l1','l2']}
cv=KFold(n_splits=5,random_state=None,shuffle=False)
clf=GridSearchCV(lr,grid,cv=cv,n_jobs=-1,scoring='f1_macro')
clf.fit(X1_train,y1_train)
```

hyper parameter tuning on logistic regression

```python
y_pred1=clf.predict(X1_test)
print(confusion_matrix(y1_test,y_pred1))
print(accuracy_score(y1_test,y_pred1))
print(classification_report(y1_test,y_pred1))
```

hyper parameter tuning on random forest algorithm

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=0, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
y_predrf=rf_random.predict(X_test)
print(confusion_matrix(y_test,y_predrf))
print(accuracy_score(y_test,y_predrf))
print(classification_report(y_test,y_predrf))
```

# Handling imbalanced Data


1-Since the Dataset was imbalanced, we performed SMOTE technique and checked whether it improves the accuracy or not.

```python
from imblearn.combine import SMOTETomek
```

```python
from collections import Counter
os=SMOTETomek(0.75)
X_train_ns,y_train_ns=os.fit_sample(X_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_ns)))
```

```python
rf1=RandomForestClassifier(random_state=0)
rf1.fit(X_train_ns,y_train_ns)
y_predr=rf1.predict(X_test)
print(confusion_matrix(y_test,y_predr))
print(accuracy_score(y_test,y_predr))
print(classification_report(y_test,y_predr))
```

```python
rf_random.fit(X_train_ns,y_train_ns)
y_predr=rf_random.predict(X_test)
print(confusion_matrix(y_test,y_predr))
print(accuracy_score(y_test,y_predr))
print(classification_report(y_test,y_predr))
```

```python
classifier = XGBClassifier(random_state=0)
classifier.fit(X_train_ns,y_train_ns)
y_predx1 = classifier.predict(X_test)
print(confusion_matrix(y_test,y_predx1))
print(accuracy_score(y_test,y_predx1))
print(classification_report(y_test,y_predx1))
```

```python
X_train_ns1,y_train_ns1=os.fit_sample(X1_train,y1_train)
print("The number of classes before fit {}".format(Counter(y1_train)))
print("The number of classes after fit {}".format(Counter(y_train_ns1)))
```

```python
clas=SVC(kernel='rbf',random_state=0)
clas.fit(X_train_ns1, y_train_ns1)
y_preds = clas.predict(X1_test)
print(confusion_matrix(y1_test,y_preds))
print(accuracy_score(y1_test,y_preds))
print(classification_report(y1_test,y_preds))
```

```python
lr=LogisticRegression(random_state=0)
lr.fit(X_train_ns1,y_train_ns1)
y_pred2=lr.predict(X1_test)
print(confusion_matrix(y1_test,y_pred2))
print(accuracy_score(y1_test,y_pred2))
print(classification_report(y1_test,y_pred2))
```

After SMOTE technique the accuracy of all the models decreased slighlty.

# Observation and Inference

1-In credit risks the focus should be to minimize False Positives – which means the actual defaulters which were wrongly classified as non-defaulters,
which could affect indirectly affect a company performance. 

2-Hence apart from accuracy, precision (which is the ratio of the accurately predicted defaulters) 
will help decide the performance of the model.

3-Observing the result obtained before and after applying smote , the model with best precision in predicting defaulters is logistic regression.
and best accuracy is SVM and XGB Classifier
